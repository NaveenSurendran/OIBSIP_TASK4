# SPLITTING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(espam.Email,espam.Category,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
X_train_count=vect.fit_transform(X_train.values)
X_train_count.toarray()
```

CountVectorizer is used to transform a given text into a vector on the basis of the frequency (count) of each word that occurs in the entire text.

```python
X_train,X_test
```

```python
y_train,y_test
```


# MULTINOMIAL NAIVE BAYES 

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_count, y_train)
```

Multinomial naive bayes is particularly useful for problems that involve text data with discrete features such as word frequency counts.

```python
x_test_count = vect.transform(X_test)
accuracy=model.score(x_test_count, y_test)
print(f'Accuracy of the model:{accuracy*100:.2f}%')
```

```python
def predict():
    message=input("enter the message to predict:")
    message=[message]
    vector = vect.transform(message)
    if (model.predict(vector)==1):
        print("SPAM EMAIL")
    else:
        print("NOT A SPAM")
predict()
```
